```python
<!DOCTYPE html>
<html lang="es">
<head>
  <meta charset="UTF-8" />
  <meta name="viewport" content="width=device-width, initial-scale=1.0"/>
  <title>Mundo Miko</title>
  <style>
    @import url('https://fonts.googleapis.com/css2?family=Fredoka+One&display=swap');
    body {
      font-family: 'Fredoka One', cursive;
      background: linear-gradient(135deg, #ffdee9, #b5fffc);
      margin: 0;
      padding: 0;
      text-align: center;
    }
    .pantalla {
      display: none;
      padding: 20px;
      animation: fadeIn 1s ease forwards;
    }
    .activa {
      display: block;
    }
    h1 {
      font-size: 42px;
      color: #ff4d6d;
      margin: 0;
      padding-top: 10px;
      display: flex;
      justify-content: center;
      align-items: center;
      gap: 15px;
      animation: bounce 2s infinite;
    }
    h1 img {
      height: 60px;
    }
    .miko {
      height: 60px;
    }
    .btn {
      background-color: #ff69b4;
      border: none;
      padding: 12px 24px;
      font-size: 20px;
      color: white;
      border-radius: 15px;
      cursor: pointer;
      transition: transform 0.3s ease;
      margin-top: 20px;
    }
    .btn:hover {
      transform: scale(1.1);
    }
    .drag-container {
      display: flex;
      justify-content: center;
      flex-wrap: wrap;
      gap: 20px;
      margin-top: 20px;
    }
    .drop-zone {
      width: 150px;
      height: 150px;
      border: 4px dashed #999;
      border-radius: 15px;
      display: flex;
      justify-content: center;
      align-items: center;
      background-color: white;
      font-size: 40px;
      transition: background-color 0.3s ease;
    }
    .drop-zone:hover {
      background-color: #f0f8ff;
    }
    .palabra {
      padding: 12px 16px;
      background-color: #8ecae6;
      border-radius: 12px;
      cursor: grab;
      font-size: 22px;
      transition: transform 0.2s ease;
    }
    .palabra:hover {
      transform: scale(1.1);
    }
    @keyframes fadeIn {
      from { opacity: 0; transform: translateY(20px); }
      to { opacity: 1; transform: translateY(0); }
    }
    @keyframes bounce {
      0%, 100% { transform: translateY(0); }
      50% { transform: translateY(-10px); }
    }
  </style>
</head>
<body>

  <div id="inicio" class="pantalla activa">
    <h1>
      <img class="miko" src="https://i.imgur.com/7YV1C8X.png" alt="Miko" />
      <img src="https://upload.wikimedia.org/wikipedia/commons/thumb/5/53/Globe_icon.svg/1024px-Globe_icon.svg.png" alt="mundo" />
      Mundo Miko
    </h1>
    <h2>Misi√≥n: Francia</h2>
    <button class="btn" onclick="mostrarPantalla('mision')">¬°Empezar!</button>
  </div>

  <div id="mision" class="pantalla">
    <h2>üá´üá∑ Bonjour! Soy Miko ‚ú®</h2>
    <p>Arrastr√° la palabra correcta al emoji correspondiente.</p>
    <div class="drag-container">
      <div class="drop-zone" data-palabra="fromage">üßÄ</div>
      <div class="drop-zone" data-palabra="chat">üê±</div>
      <div class="drop-zone" data-palabra="bonjour">üëã</div>
    </div>
    <div class="drag-container" style="margin-top: 30px;">
      <div class="palabra" draggable="true" id="fromage">fromage</div>
      <div class="palabra" draggable="true" id="bonjour">bonjour</div>
      <div class="palabra" draggable="true" id="chat">chat</div>
      <div class="palabra" draggable="true" id="chien">chien</div>
      <div class="palabra" draggable="true" id="merci">merci</div>
      <div class="palabra" draggable="true" id="pomme">pomme</div>
    </div>
  </div>

  <div id="victoria" class="pantalla">
    <h2>üéâ ¬°Misi√≥n completada! üéâ</h2>
    <p>¬°Felicitaciones! Aprendiste nuevas palabras en franc√©s üá´üá∑</p>
  </div>

  <script>
    function mostrarPantalla(id) {
      document.querySelectorAll('.pantalla').forEach(p => p.classList.remove('activa'));
      document.getElementById(id).classList.add('activa');
    }

    let aciertos = 0;

    document.querySelectorAll('.palabra').forEach(elem => {
      elem.addEventListener('dragstart', e => {
        e.dataTransfer.setData('text', e.target.id);
      });
    });

    document.querySelectorAll('.drop-zone').forEach(zone => {
      zone.addEventListener('dragover', e => e.preventDefault());
      zone.addEventListener('drop', e => {
        const id = e.dataTransfer.getData('text');
        const dragged = document.getElementById(id);
        if (id === zone.dataset.palabra) {
          zone.textContent = dragged.textContent + ' ‚úîÔ∏è';
          dragged.remove();
          aciertos++;
          if (aciertos === 3) {
            setTimeout(() => mostrarPantalla('victoria'), 700);
          }
        } else {
          alert('Oops, esa no es la palabra correcta.');
        }
      });
    });
  </script>

</body>
</html>
```
